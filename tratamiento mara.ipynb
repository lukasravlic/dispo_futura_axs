```python
mara = pd.read_excel("C:/Users/lravlic/Inchcape/Planificación y Compras Chile - Documentos/Planificación y Compras Maestros/2024/2024-08/MARA_R3_20240801.xlsx")
```

```python
# Función para modificar 'Material' en función del valor de 'Tipo'
def modificar_material(row):
    if row['Material'].endswith('INP300'):
        return row['Material'][:-6]  # Elimina los últimos 2 caracteres
    elif row['Material'].endswith('R3'):
        return row['Material'][:-2]  # Elimina los últimos 6 caracteres
    else:
        return row['Material']  # No modifica si no cumple ninguna condición

# Aplicar la función a la columna 'Material'
mara['Material'] = mara.apply(modificar_material, axis=1)


```

```python
mara.drop_duplicates(subset='Material_R3', keep='first', inplace=True)
```
